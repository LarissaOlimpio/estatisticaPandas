```python
import pandas as pd
import statistics as st

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('casos_obitos_doencas_preexistentes.csv', sep=';',encoding='cp1252')

df.head(20)
```

```python
df = pd.read_csv('casos_obitos_doencas_preexistentes.csv', sep=';',encoding='cp1252')
df.drop('asma', inplace=True, axis=1)
df.drop('cardiopatia', inplace=True, axis=1)
df.drop('diabetes', inplace=True, axis=1)
df.drop('doenca_hematologica', inplace=True, axis=1)
df.drop('doenca_hepatica', inplace=True, axis=1)
df.drop('doenca_neurologica', inplace=True, axis=1)
df.drop('doenca_renal', inplace=True, axis=1)
df.drop('imunodepressao', inplace=True, axis=1)
df.drop('obesidade', inplace=True, axis=1)
df.drop('outros_fatores_de_risco', inplace=True, axis=1)
df.drop('pneumopatia', inplace=True, axis=1)
df.drop('puerpera', inplace=True, axis=1)
df.drop('sindrome_de_down', inplace=True, axis=1)
```

```python
st.median(df.idade) #A Mediana
#representa o valor central de um conjunto de dados.
#Nesse caso em relação a idade a mediana é 47
```

```python
st.mode(df.nome_munic)#MODA
# A Moda representa o valor mais frequente de um conjunto de dados, sendo assim, para defini-la basta observar a 
#frequência com que os valores aparecem, sendo assim neste caso o valor que mais aparareceu foi São Paulo.
```

```python
df = pd.read_csv('Dados-covid-19-municipios(1).csv', sep=';',encoding='cp1252')
#Relação de municipios e quantidade de casos e total de óbito devido ao covid-19
df.head(20)
```

```python
df.columns
```

```python
st.mean(df.Mun_Totaldeobitos) #media
#A média é calculada somando-se todos os valores de um conjunto de dados
#e dividindo-se pelo número de elementos deste conjunto.
#neste caso temos a media de números de casos de covid-19 no estado de São Paulo que chegaram a óbito 
```

```python
st.stdev(df.Mun_Totaldeobitos)# desvio padrão
#O desvio padrão é uma medida que expressa o grau de dispersão de um conjunto de dados. Ou seja, 
# indica o quanto um conjunto de dados é uniforme.
#Quanto mais próximo de 0 for o desvio padrão, mais homogêneo são os dados.
#Neste caso indica que os dados estão espalhados por uma ampla gama de valores.
```

```python
st.variance(df.Mun_Totaldeobitos)#variancia
#Variância é uma medida de dispersão e é usada também para expressar o quanto um conjunto de dados se desvia da média.
```
